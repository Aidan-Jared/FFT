```python
import polars as pl
```

```python
lf = pl.scan_csv("../data/gpt2_rfft.csv")
```

```python
col = "phase"
lf_amplitude = lf.select(
    pl.col("Tokens"), pl.col(col)
    ).with_columns(
        pl.col(col)
        .str.replace_all(r"\[|\]", "")
        .str.replace(",", "")
        .str.strip_chars()
        .str.replace_all(r"\s+", " ")
        .str.split(" ")
        .cast(pl.List(pl.Float32))
        .list.to_struct(upper_bound=385)
        ).unnest(col)
```

```python
var = "field_1"
lf_amplitude.select(["Tokens", var]).unique().top_k(10, by=var).collect()
```
